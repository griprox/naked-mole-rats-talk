```python
import os
import sys
path_to_this_notebook = os.path.abspath('.')
PATH_TO_PROJECT = path_to_this_notebook[: path_to_this_notebook.find('notebooks')]
sys.path.append(PATH_TO_PROJECT)

import numpy as np
import re
import pandas as pd
```

```python
path_to_recordings_metadata = PATH_TO_PROJECT + 'data/'
metadata = pd.read_csv(path_to_recordings_metadata + 'recordings_metadata.csv')
```

```python
new_cols = ['split', 'labeled', 'traced', 'splits checked',  'labels checked', 'traces checked']
```

```python
rec_name
```

```python
new_cols_vals = []

for rec_name, rec_path, stage in metadata[['name', 'path', 'processing stage']].values:
    try:
        rec_split = 'auto'
        rec_labeled = False
        rec_traced = False
        
        splits_checked = False
        labels_checked = False
        traces_checked = False
        
        txt = pd.read_csv(rec_path + rec_name.replace('npy', 'txt'), sep='\t')
        sound_values = txt['cl'].unique()
        if len(sound_values) == 0:
            new_cols_vals.append((rec_split, rec_labeled, rec_traced,
                                  splits_checked, labels_checked, traces_checked))
            continue
        
        elif not ('sound' in sound_values and 
                  ((len(sound_values) == 1) or (len(sound_values) == 2 and 'noise' in sound_values))):
            rec_labeled = 'auto'
            
        if stage == 'split and checked':
            splits_checked = True
        elif stage == 'labeled and checked':
            splits_checked = True
            labels_checked = True
            
        traces_in_rec = [t[:t.find('_')] for t in os.listdir(rec_path.replace('recordings', 'traces')) 
                         if (rec_name[:-3] in t) and 'trace' in t]
        if len(traces_in_rec) > 0:
            rec_traced = 'auto'
            if 'freshtrace' not in traces_in_rec:
                traces_checked = True
                
                
        
    except:
        rec_split = False
        rec_labeled = False
        rec_traced = False
        splits_checked = False
        labels_checked = False
        traces_checked = False
    if rec_labeled == False and labels_checked:
        print(123)
        break
    new_cols_vals.append((rec_split, rec_labeled, rec_traced, 
                          splits_checked, labels_checked, traces_checked))
new_cols_vals = np.array(new_cols_vals)
```
