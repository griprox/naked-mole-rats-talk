```python
import os
import sys
path_to_this_notebook = os.path.abspath('.')
PATH_TO_PROJECT = path_to_this_notebook[: path_to_this_notebook.find('notebooks')]
sys.path.append(PATH_TO_PROJECT)

import numpy as np
import pandas as pd
import os
import pickle
import matplotlib.pyplot as plt
from src.data_representations.softchirp_features import *
from src.preprocessing.filters import filter_recs_metadata
from src.preprocessing.process_sounds_metadata import generate_sounds_metadata
from src.preprocessing.load_data import load_recs_dict
from src.data_representations.process_wavs import *
from src.data_representations.process_images import *
from src.deep_learning.util import *
import re

from src.preprocessing.filters import filter_recs_metadata, filter_sounds_metadata_extended
from src.visualization.plotting_utils import plot_confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

from notebooks.analysis.plotting_config import ORDER, COLOR
from sklearn.ensemble import RandomForestClassifier
from src.visualization.plot_softchirps import plot
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA, KernelPCA
from sklearn.manifold import TSNE
import umap

%matplotlib inline


def get_features_to_use(features, features_filter, projection_method,
                        normalize=False, **projection_kwargs):
    if features_filter == 'all':
        features_to_use = features
    elif features_filter == 'mfcc':
        features_to_use = features[[c for c in features.columns if 'mfcc' in c]]
    else:
        features_to_use = features[features_filter]
    features_to_use = pd.DataFrame(StandardScaler().fit_transform(features_to_use), columns=features_to_use.columns)
    if projection_method in ['PCA', 'pca']:
        transform = KernelPCA(n_components=2, **projection_kwargs)
        features_to_use = transform.fit_transform(features_to_use)
        cols = ['Dim#1', 'Dim#2']
    elif projection_method in ['TSNE', 'tsne', 'tSNE']:
        transform = TSNE(n_components=2, **projection_kwargs)
        features_to_use = transform.fit_transform(features_to_use)
        cols = ['Dim#1', 'Dim#2']
    elif projection_method in ['UMAP', 'umap', 'uMAP']:
        transform = umap.UMAP(n_components=2, **projection_kwargs)
        features_to_use = transform.fit_transform(features_to_use)
        cols = ['Dim#1', 'Dim#2']
    else:
        cols = features_to_use.columns
    return pd.DataFrame(features_to_use, columns=cols)


def load_sounds_extract_features(sounds_metadata, recs_metadata, recs_dict=None, 
                                 sr=22050, win_length=220, n_mfcc=30, denoise=False, **denoising_params):
    features = [None for _ in range(sounds_metadata.shape[0])]
    rec_names_set = sounds_metadata['rec'].unique()
    if recs_dict is None:
        recs_dict = {}
    for rec_numb, rec_name in enumerate(rec_names_set):
        print('Loading and processing recording %d/%d' % (rec_numb + 1, len(rec_names_set)))
        rec_path = recs_metadata[recs_metadata['name'] == rec_name]['path'].iloc[0]
        if rec_name in recs_dict:
            rec = recs_dict[rec_name]  
        else:
            print('Loaded')
            rec = np.load(rec_path + rec_name)
            rec = denoise_recording(rec, **denoising_params) if denoise else rec
            
        rec_sounds_inds = np.where(sounds_metadata['rec'] == rec_name)[0]
        sounds_metadata_in_rec = sounds_metadata[sounds_metadata['rec'] == rec_name]
        features_sounds_in_rec = extract_features(sounds_metadata_in_rec, rec,  sr, win_length, n_mfcc)
        for ind_loc, ind in enumerate(rec_sounds_inds):
            features[ind] = np.copy(features_sounds_in_rec.iloc[ind_loc])
        columns = features_sounds_in_rec.columns
        del(features_sounds_in_rec)
        del(rec)
    features = pd.DataFrame(features, columns=columns)
    return features
    

def extract_features(sounds_metadata, rec, sr=22050, win_length=220, n_mfcc=30):
    features = []
    for s, e in sounds_metadata[['s', 'e']].values:
        s_int, e_int = int(sr * s), int(sr * e)
        sound = rec[s_int: e_int]
        
        pitch = compute_pitch(sound, win_length)
        ampl = compute_amplitude(sound, win_length)
        ent = compute_wiener_ent(sound, win_length)
        zero_cross = compute_zero_crossing(sound)
        duration_full = e - s
        
        mfcc_features = librosa.feature.mfcc(sound, n_mfcc=n_mfcc)
        mfcc_features = np.concatenate([mfcc_features.min(1), mfcc_features.mean(1), mfcc_features.max(1)])
        
        all_features = np.stack([pitch, ampl, ent, zero_cross, zero_cross] + mfcc_features.tolist())
        features.append(all_features)
    return pd.DataFrame(features, columns=['pitch', 'amplitude', 'Ã©ntropy', 'zero crossing', 'duration'] + 
                        np.reshape([('mfcc_min%d' % i, 'mfcc_mean%d' % i, 'mfcc_max%d' % i) for i in range(30)], 
                                   (-1)).tolist())
        
```

```python
path_to_recordings_metadata = PATH_TO_PROJECT + 'data/'
recordings_metadata_name = 'recordings_metadata.csv'

# what data to load
dates = 'all'
colonies = 'all'
#colonies = ['martell', 'arryn', 'targaryen', 'lannister']
colonies = ['dam', 'nmr', 'hvd']
experiments = 'all'
stages = 'all'
sr = 22050
preload_recs_dict = False
minimal_sound_length_sr = 2048
denoise = False
denoising_params = {'sr': sr,
                    'stationary':False,
                    'n_fft': 256, 
                    'time_constant_s': 2, 
                    'freq_mask_smooth_hz': 500, }
classes_to_drop = ['noise', 'RFIDscanner', 'weirdo', 'multiplesoftchirps', 'badsplit']

# load metadata
recs_metadata = pd.read_csv(path_to_recordings_metadata + recordings_metadata_name)
mask = filter_recs_metadata(recs_metadata, dates, colonies, stages, experiments)
recs_metadata = recs_metadata[mask]
recs_metadata = recs_metadata.reset_index(drop=True)

print('Training recs: %d' % len(recs_metadata))

sounds_metadata = generate_sounds_metadata(recs_metadata).reset_index(drop=True)
print('%d sounds in total' % len(sounds_metadata))
sounds_metadata = sounds_metadata[sounds_metadata['e']-sounds_metadata['s']>= minimal_sound_length_sr/sr].reset_index(drop=True)
print('%d sounds after filtering' % len(sounds_metadata))
classes_to_drop = classes_to_drop + [cl for cl in sounds_metadata['cl'].unique() if 'multipl' in cl]
sounds_metadata = sounds_metadata[~sounds_metadata['cl'].isin(classes_to_drop)].reset_index(drop=True)
print('%d sounds after filtering' % len(sounds_metadata))
# Preaload recs dict, if needed
print('Loaded sounds metadata, shape is %s' % (sounds_metadata.shape, ))
if preload_recs_dict:
    print('Loading recs dict . . .')
    recs_dict = {}
    for rec_name in sounds_metadata['rec'].unique():
        rec_path = recs_metadata[recs_metadata['name'] == rec_name]['path'].iloc[0]
        rec = np.load(rec_path + rec_name)
        recs_dict[rec_name] = denoise_recording(rec, **denoising_params) if denoise else rec
    print('Done !')
else:
    recs_dict = {}
```

```python
file_name = 'features_%s' % ''.join([c[:1].upper() for c in colonies])
file_name = file_name + '_denoise' if denoise else file_name
print(PATH_TO_PROJECT + 'features_csvs/' + file_name)

# Compute features
win_length = 220
n_mfcc = 30

if os.path.isfile(PATH_TO_PROJECT + 'features_csvs/' + file_name):
    features_ext = pd.read_csv(PATH_TO_PROJECT + 'features_csvs/' + file_name)
else:
    features = load_sounds_extract_features(sounds_metadata, recs_metadata, recs_dict=recs_dict,
                                            sr=sr, win_length=win_length, n_mfcc=n_mfcc)
    features_ext = pd.concat([sounds_metadata, features], axis=1)
    del(features)
    # Save
    features_ext.to_csv(PATH_TO_PROJECT + 'features_csvs/' + file_name, index=None)
    
metadata_cols = ['s', 'e', 'cl', 'rec', 'colony', 'ratids', 'date', 'experiment']
feature_cols = [c for c in features_ext.columns if c not in metadata_cols]

sounds_metadata = features_ext[metadata_cols]
features = features_ext[feature_cols]
```

# Analysis

```python
# Set filters
colonies = ['hvd']
weights = 'all'
dates = 'all'
sexes = 'all' 
ages = 'all'
ranks = 'all' 
ratids = 'all'
excludeids = None 
bodylengths = 'all'
bodylengths2 = 'all' 
sounds_per_rat = [0, np.inf]
types = 'all'

classes_to_plot = 'all'
classes_to_drop = ['sound',]

mask =  filter_sounds_metadata_extended(sounds_metadata, ratids, excludeids, 
                                   types, colonies, weights, sexes, ages, ranks, dates,
                                   bodylengths, bodylengths2, sounds_per_rat)
if classes_to_plot != 'all':
    mask *= sounds_metadata['cl'].isin(classes_to_plot)
if classes_to_drop:
    mask *= ~sounds_metadata['cl'].isin(classes_to_drop)
    
```

```python
# Choose features and get projection
features_filter = 'all' # 'all', 'mfcc' or list of features
#features_filter = ['pitch', 'duration']
projection_method = 'tsne'
#projection_method = None
all_projections_kwargs = {'pca': {'kernel': 'linear'}, 
                          'tsne': {'perplexity': 30.0}, 
                          'umap': {'min_dist': .1, 'metric': 'euclidean', 'n_neighbors': 20},
                          None: {}}

sounds_metadata_to_use = sounds_metadata[mask].reset_index(drop=True)

features_to_use = features[mask].reset_index(drop=True)
features_to_use = get_features_to_use(features_to_use, features_filter, projection_method,
                                      normalize=True, **all_projections_kwargs[projection_method])
print(features_to_use.shape)
```

```python
clf = RandomForestClassifier(n_estimators=128)
X_train, X_test, y_train, y_test = train_test_split(features_to_use, y_raw, test_size=0.33)
len(set(y_test))
```
