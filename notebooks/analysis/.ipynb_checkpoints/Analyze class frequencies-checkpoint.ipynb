```python
# specify paths, where data is stored
path_to_info = PATH_TO_PROJECT + 'data/rat_info.csv'
path_to_recordings_metadata = PATH_TO_PROJECT + 'data/'
recordings_metadata_name = 'recordings_metadata.csv'

# what data to load
dates = 'all'
colonies = ['hvdkleinannotated','nmrkleinannotated','damkleinannotated','bsuilkleinannotated',]#['hvdkleinannotated','nmrkleinannotated','damkleinannotated','bsuilkleinannotated']
experiments = 'all'# 'hvdkleinannotated','nmrkleinannotated','damkleinannotated','bsuilkleinannotated'
experiments = 'all'
stages = ['labeled and checked', 'traced and checked', 'labeled', 'traced']


# Modify when you add a new colony
colonies_dict = {'nmrkleinannotated': 'nmr', 
                 'damkleinannotated': 'dam',
                 'hvdkleinannotated': 'hvd',
                 'bsuilkleinannotated': 'bsuil'}

# In which order colonies will be plotted. Modify when you add a new colony
plotting_order = ['hvdkleinannotated', 'damkleinannotated', 'nmrkleinannotated', 'bsuilkleinannotated']
```

```python
info = load_info_file(path_to_info)

recs_metadata = pd.read_csv(path_to_recordings_metadata + recordings_metadata_name)
mask = filter_recs_metadata(recs_metadata, dates, colonies, stages, experiments)
recs_metadata = recs_metadata[mask].reset_index(drop=True)
    
sounds_metadata = generate_sounds_metadata(recs_metadata, [], {}, 0, np.inf)
txts_dict = {}
for rec_name, rec_path in recs_metadata[['name', 'path']].values:
    txt = pd.read_csv(rec_path + rec_name.replace('npy', 'txt'), sep='\t')
    txt =  txt.sort_values(by='s').reset_index(drop=True)
    txts_dict[rec_name] = txt
```

```python
# Find pairs of bsuil recs which we want to merge
recordings_to_merge = set()
renaming_dict = {}
all_bsuil_recs = recs_metadata[recs_metadata['colony'] == 'bsuilkleinannotated']['name'].values.tolist()
get_number = lambda x:x[-12:-4]
for rec_name in all_bsuil_recs:
    n = get_number(rec_name)
    if n[0] == '2':
        n_pair = '1' + n[1:]
    elif n[0] == '1':
        n_pair = '2'  + n[1:]
    else:
        n_pair = None
    
    rec_pair_name = [rn for rn in all_bsuil_recs if get_number(rn) == n_pair][0]
    recordings_to_merge.add(tuple(sorted((rec_name, rec_pair_name))))

    new_name = sorted((rec_name, rec_pair_name))[0][:-4] + 'merge.npy'
    renaming_dict[rec_name] = new_name
    renaming_dict[rec_pair_name] = new_name
    
recordings_to_merge = list(recordings_to_merge)
# Update sounds metadata and txt_dict based on recordings_to_merge
sounds_metadata['rec'] = sounds_metadata['rec'].apply(lambda x: renaming_dict[x] if x in renaming_dict else x)
txts_dict_merge = {}
for rec_name, rec in txts_dict.items():
    if rec_name not in renaming_dict:
        txts_dict_merge[rec_name] = rec
        
for rn1, rn2 in recordings_to_merge:
    new_name = renaming_dict[rn1]
    txt1 = txts_dict[rn1]
    txt2 = txts_dict[rn2]
    txt_merge = pd.concat([txt1, txt2]).sort_values(by='s').reset_index(drop=True)
    txts_dict_merge[new_name] = txt_merge
txts_dict = txts_dict_merge
```

```python
multi_rep_dict = {'multiplesoftchirps': ('sc', 2), 'multiplesoftchirps2': ('sc', 2), 
                 'multiplesoftchirps3': ('sc', 3), 'multiplesoftchirps4': ('sc', 4), 
                 'multiplesoftchirps5': ('sc', 5), 'multiplesoftchirps6': ('sc', 6),
                 'multiplesoftchirps7': ('sc', 7), 
                  
                 'multipleloudchirp': ('lc', 2), 'multipleloudchirps2': ('lc', 2),
                 'multipleloudchirps3': ('lc', 3), 'multipleloudchirps4': ('lc', 4),
                 'multipleloudchirps5': ('lc', 5), 'multipleloudchirps6': ('lc', 6),
                 'multipleloudchirps7': ('lc', 7),
                  
                 'multiplecombos': ('cm', 2), 'multiplecombos2': ('cm', 2),
                 'multiplecombos3': ('cm', 3), 'multiplecombos4': ('cm', 4),
                 'multiplecombos5': ('cm', 5), 'multiplecombos6': ('cm', 6), 
                 'multiplecombos7': ('cm', 7),
                 
                 'multiplewhistles2': ('wh', 2), 'multiplewhistles3': ('wh', 3), 'multiplewhistles4': ('wh', 4),
                 
                 }

#group_dict = {'softchirp': 'chirp', 'loudchirp': 'chirp'} 'class_name': 'group name'
group_dict = {}

class_to_multi = defaultdict(list)
for key, (cl, _) in multi_rep_dict.items():
    if cl == 'sc':
        cl = 'softchirp'
    elif cl == 'lc':
        cl = 'loudchirp'
    elif cl == 'cm':
        cl = 'combo'
    class_to_multi[cl].append(key)
```

```python
unsplit = {'tweet1combo1': [('tw', 1), ('cm', 1)],
           'downsweep1cheveron1': [('ds', 1), ('ch', 1)], 
           'loudchirp1combo1': [('lc', 1), ('cm', 1)], 
           'mordent2': [('mr', 2)], 
           'downsweep2': [('ds', 2)], 
           'loudchirp2softchirp1': [('lc', 2), ('sc', 1)],
           'vtrill2': [('vtrill', 2)],
           'downsweep3': [('ds', 3)],
           'whistle2softchirp3': [('wh', 2), ('sc', 3)],
           'loudchirp1combo2': [('lc', 1), ('cm', 2)] ,
           'combo1softchirp1': [('cm', 1), ('sc', 1)],
           'whistle1softchirp1': [('wh', 1), ('sc', 1)],
           'whistle1softchirp2': [('wh', 1), ('sc', 2)],
           'upsweep2': [('us', 2)],
           'loudchirp1softchirp1': [('lc', 1), ('sc', 1)],
           'squawk2': [('squawk', 2)], 
           'phee2': [('ph', 2)],
           'tweet2': [('tw', 2)],
           'whistle3softchirp2': [('wh', 3), ('sc', 2)],
           'whistle2softchirp2': [('wh', 2), ('sc', 2)],
           'whistle2softchirp1': [('wh', 2), ('sc', 1)],
           'phee1softchirp1': [('ph', 1), ('sc', 1)],
           'squawk2mordent1': [('sq', 2), ('mr', 1)],
           'softchirp1squawk1': [('sc', 1), ('sq', 1)],
           'loudchirp1whistle1': [('lc', 1), ('wh', 1)],
           'whistle2loudchirp2': [('wh', 2), ('lc', 2)],
}
```

```python
# This cell is slow !!! 
total_durations = defaultdict(lambda: 0)
recs_durations = dict()
for rec_name, rec_path, colony in recs_metadata[['name', 'path', 'colony']].values:
    rec = np.load(rec_path + rec_name)
    duration = rec.shape[0] / 22050
    print('%s colony %s lasts for %.2f minutes' % (colony, rec_name, duration / 60))
    if colony == 'bsuilkleinannotated':
        rec_name = renaming_dict[rec_name] if rec_name in renaming_dict else rec_name
        if rec_name in recs_durations:
            continue        
    total_durations[colony] += duration
    recs_durations[rec_name] = duration
    del(rec)
    
for col, dur in total_durations.items():
    print('Total duration of recordings from %s is %.2f minutes' % (col, dur/60))
```

```python
not_sound_classes = ['badsplit','noise', 'ns','sx',] # Specifies what sounds are counted for the normalization
#classes_to_plot = ['softchirp', 'loudchirp', 'phee', 'tweet'] # What classes to plot
classes_to_plot = 'all'
classes_to_drop = ['cheveronphee', 'weirdo', 'sound',]
if classes_to_drop is None:
    classes_to_drop = []
all_notmulti_classes = set([cl for cl in sounds_metadata['cl'].unique() if 'multi' not in cl])
unsplit_classes = set([cl for cl in sounds_metadata['cl'].unique() if cl in unsplit])
sound_classes = set(sounds_metadata['cl'].unique()) - set(not_sound_classes)
if classes_to_plot == 'all':
    classes_to_plot = (set(all_notmulti_classes) - set(unsplit_classes) - 
                       set(not_sound_classes) - set(classes_to_drop))
classes_to_consider = set(sounds_metadata['cl'].unique()) - set(not_sound_classes)

per_colony_class_occurences = defaultdict(lambda: defaultdict(lambda: 0))
per_colony_class_frequency = defaultdict(lambda: defaultdict(list))
#classes_to_plot = set()

for col in sounds_metadata.colony.unique():
    #for col in ['bsuilkleinannotated']:   
    col_mask = sounds_metadata['colony'] == col
    sounds_metadata_col = sounds_metadata[col_mask].reset_index(drop=True)
    for rec_name in sounds_metadata_col['rec'].unique():
        txt = txts_dict[rec_name]
        duration = recs_durations[rec_name]
        cl_occured_in_txt = defaultdict(lambda: 0)
        for ind, (s, e, cl) in enumerate(txt.values):
            if cl not in classes_to_consider:
                continue
            if 'multi' in cl:
                cl, rep = multi_rep_dict[cl]
                if cl == 'sc':
                    cl = 'softchirp'
                elif cl == 'lc':
                    cl = 'loudchirp'
                elif cl == 'cm':
                    cl = 'combo'
            else:
                rep = 1
            if cl in group_dict:
                cl = group_dict[cl]
            per_colony_class_occurences[col][cl] += rep
            if cl in unsplit:
                #print(cl)
                for cl_i, rep in unsplit[cl]:
                    #print(cl_i, rep)
                    cl_occured_in_txt[cl_i] += rep
            else:
                cl_occured_in_txt[cl] += rep
            #classes_to_plot.add(cl)
        for cl in cl_occured_in_txt:
            per_colony_class_frequency[col][cl].append(60 * cl_occured_in_txt[cl] / duration)
            #print(rec_name, 60 * cl_occured_in_txt[cl] / duration)
            
ordering_function = lambda x: ORDER['cl'].index(x) if x in ORDER['cl'] else np.inf
classes_to_plot = sorted(classes_to_plot)
```

```python
# Plot class occurences
plt.figure(figsize=(20, 6))
normalize = False

colonies_to_use = ['hvdkleinannotated', 'damkleinannotated','nmrkleinannotated', 'bsuilkleinannotated']
colonies_to_use = sorted(colonies_to_use, key=lambda x: plotting_order.index(x))
n_classes = len(classes_to_plot)
n_colonies = len(per_colony_class_frequency.keys())
x = np.arange(0, n_classes * (n_colonies + 1), 1)
x_ticks = np.arange((n_colonies - 1)/2, n_classes * (n_colonies + 1), n_colonies + 1)
classes_occurences = []
colonies_ordered = []

for col in colonies_to_use:
    col_ind = colonies.index(col)
    colonies_ordered.append(col)
    x_col = x[col_ind::(n_colonies + 1)]
    y_col = np.array([per_colony_class_occurences[col][cl] for cl in classes_to_plot])
    if normalize:
        y_col = y_col / sum(per_colony_class_occurences[col].values())
    classes_occurences.append(y_col)
    c = COLOR['colony'][colonies_dict[col]]
    if c is not None:
        plt.bar(x_col, y_col, label=colonies_dict[col], color=c)
    else:
        plt.bar(x_col, y_col, label=colonies_dict[col])
_ = plt.legend()

plt.title('Class occurences per colony')
if normalize:
    plt.ylabel('Ratio of sound over all sounds')
else:
    plt.ylabel('Number of occurences')
    
_ = plt.xlabel('Class')
_ = plt.xticks(x_ticks, list(classes_to_plot))

df = pd.DataFrame(np.transpose(classes_occurences), columns=colonies_ordered,)
df.index = classes_to_plot
# Uncomment line below to save
df.to_csv('class occurences_all_dam_nmr_hvd_busil_coloniesof4_realvalues_withoutsoftchirps_23apr2022.csv',)
#plt.savefig("class occurences_all_dam_nmr_hvd_busil_coloniesof4_realvalues_withoutsoftchirps_05apr2022_2.svg", format="svg")
df
```

```python

# Plot class frequencies
plt.figure(figsize=(20, 6))
normalize_by_max = True

colonies_to_use = ['damkleinannotated','hvdkleinannotated', 'bsuilkleinannotated','nmrkleinannotated']
colonies_to_use = sorted(colonies_to_use, key=lambda x: plotting_order.index(x))

gs = gridspec.GridSpec(1, 2, width_ratios=[3, 1]) 
ax0 = plt.subplot(gs[0])
n_classes = len(classes_to_plot)
n_colonies = len(per_colony_class_frequency.keys())
x = np.arange(0, n_classes * (n_colonies + 1), 1)
x_ticks = np.arange((n_colonies - 1)/2, n_classes * (n_colonies + 1), n_colonies + 1)
total_frequency_per_colony = {}

classes_frequencies = []
colonies_ordered = []


for col_ind, col in enumerate(colonies_to_use):
    # col_ind = plotting_order.index(col)
    colonies_ordered.append(col)
    x_col = x[col_ind::(n_colonies + 1)]
    empty_mean = lambda x: np.mean(x) if len(x) else 0
    y_col = np.array([empty_mean(per_colony_class_frequency[col][cl]) for cl in classes_to_plot])
    classes_frequencies.append(y_col)    
    c = COLOR['colony'][colonies_dict[col]]
    if c is not None:
        plt.bar(x_col, y_col, label=colonies_dict[col], color=c)
    else:
        plt.bar(x_col, y_col, label=colonies_dict[col])
    total_frequency_per_colony[col] = np.nansum(y_col)
_ = plt.legend()

plt.title('Class frequencies per colony')

plt.ylabel('Number of occurences (#/minute)')
_ = plt.xlabel('Class')
_ = plt.xticks(x_ticks, list(classes_to_plot))
ax1 = plt.subplot(gs[1])

#plt.ylim([0, 5])

y = [total_frequency_per_colony[col] for col in colonies_to_use]
if normalize_by_max:
    y = [yi/max(y) for yi in y]

for col_ind, col in enumerate(colonies_to_use):
    # col_ind = colonies.index(col)
    c = COLOR['colony'][colonies_dict[col]]
    if c is not None:
        plt.bar(colonies_dict[col], y[col_ind], color=c, label=colonies_dict[col])
    else:
        plt.bar(colonies_dict[col], y[col_ind], label=colonies_dict[col])
    
    
_ = plt.title('Aggregated frequency of sounds')
plt.ylabel('Number of occurences (#/minute)')
_ = plt.xlabel('Colony')

df = pd.DataFrame(np.transpose(classes_frequencies), columns=colonies_ordered,)
df.index = classes_to_plot
# Uncomment line below to save
df.to_csv('freq occurences_all_dam_nmr_hvd_busil_coloniesof4_normalizedtomax_23apr20222_2.csv',)
#plt.savefig('freq occurences_all_dam_nmr_hvd_busil_coloniesof4_normalizedtomax_05apr2022_2.svg', format="svg")
df
```
