### Load data

```python

rec_names = [rn for rn in os.listdir(path) if '.npy' in rn]

txts = {rn: pd.read_csv(path + rn.replace('npy', 'txt'), sep='\t') for rn in rec_names}
n: np.load(path+rn) for rn in rec_names}
```

```python
sr = 250000
path_to_recordings = PATH_TO_PROJECT + '/mice_data_examples/mice/recordings/'
path_to_samples = PATH_TO_PROJECT + '/mice_samples/'
path_to_traces = PATH_TO_PROJECT + '/manually traced mouse examples/'
trace_names = os.listdir(path_to_traces)
rec_names = os.listdir(path_to_recordings)

recs_dict = {}
samples_dict = {}
sample_to_rec = {}

for rn in rec_names:
    if '.npy' not in rn:
        continue
    rec = np.load(path_to_recordings + rn)
    recs_dict[rn] = rec
    for sample in os.listdir(path_to_samples + rn[:-4]):
        sample_to_rec[sample[:-4]] = rn
        s,e = sample[:-4].split('-')
        s = int(float(s) * sr)
        e = int(float(e) * sr)
        samples_dict[sample] = rec[s : e]

```

```python
all_images = []
for sn, sound in samples_dict.items():
    im = make_image(sound)
    all_images.append(im)
all_images_rs = np.array([resize((im), (128, 64)) for im in all_images])
```

```python
images_manual, traces_manual = [], []

for sn in trace_names:
    rn = sample_to_rec[sn[:-4]]
    sound = samples_dict[sn.replace('txt', 'png')]
    
    im = make_image(sound)
    trace = np.zeros(im.shape)
    txt = np.loadtxt(path_to_traces + '/%s.txt' % sn[:-4], dtype = 'int')[:, [1, 0]]
    inds_to_delete = np.where((txt[:, 0] >= im.shape[0]) + (txt[:, 1] >= im.shape[1]))[0]
    txt = np.delete(txt, inds_to_delete, 0)
    trace[tuple(txt.T)] = 1
    images_manual.append(im)
    traces_manual.append(trace)
```

```python
Xraw = images_manual
yraw = traces_manual

X = np.array([resize(im, (128, 64)) for im in Xraw])
y = np.array([resize(im, (128, 64)) for im in yraw])
```

### Train

```python
times_aug = 7
X_aug = []
y_aug = []

for xi, yi in zip(X, y):
    for _ in range(times_aug):
        xi_aug, yi_aug = roll_pair_of_ims(xi, yi)
        X_aug.append(xi_aug)
        y_aug.append(yi_aug)
X_aug = np.array(X_aug)
y_aug = np.array(y_aug)
```

```python


input_img = Input(shape = (128, 64, 1))
x = Conv2D(64, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
encoded = MaxPooling2D((2, 2), padding='same')(x)

x = Conv2D(64, (3, 3), activation='relu', padding='same')(encoded)
x = UpSampling2D((2, 2))(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
x = Conv2D(64, (3, 3), activation='relu', padding ='same')(x)
x = UpSampling2D((2, 2))(x)
decoded = Conv2D(1, (3, 3), activation=None, padding='same')(x)

autoencoder = tf.keras.Model(input_img, decoded)
autoencoder.compile(optimizer='adadelta', loss='mean_squared_error')

model = tf.keras.Model(input_img, decoded)
model.compile(loss='mean_squared_error', optimizer = tf.keras.optimizers.Adam(lr = 4e-3))

model.summary()
```

```python
X_aug = np.reshape(X_aug, (-1, 128, 64, 1))
y_aug = np.reshape(y_aug, (-1, 128, 64, 1))
```

```python
model.fit(X_aug, y_aug, batch_size = 64, epochs = 500)
```
